#아래는 위에 입력한 '법인회생(예)'이라는 단어를 무슨 단어로 바꿀 것인지를 정의했습니다.
각 리스트에서 하나씩 빼 와서 하나의 말뭉치를 만드는 과정입니다.
만약 지역에 '동탄'을 추가하고 싶으면, 리스트 첫번째 줄에 추가해주세요.
비슷한 느낌이 드는 단어들에 추가해주시면 됩니다.

```python
import random

lists =  [
    ['', '대구', '대전', '부산', '창원', '울산', '광주', '서울', '고양', '일산', '구리', '수원', '대전', '인천', '안산', '마포', '천안아산'],
    ['기업파산', '법인파산', '법인회생', '법인회생파산' ], #법인파산 다른 이름
    ['부실기업', '책임경영'], #기업상태
    ['대표이사의 연대보증', '등기이사의 연대보증', '관재인으로서', '투자자책임'], #직책
    ['진술서', '경위서', '통지서', '선고결정문', '파산선고결정문', '파산종결결정문', '파산폐지결정문' ], #준비할 서류
    ['채무 소멸하려면', '채무변제 방법은', '개인채무 해결은', '종합해결책 비용은', '신청자격은', '예납금 면책하려면'], #원하는 결과
]

# 모든 가능한 문장들을 담을 리스트
requirement_words = []

# 모든 가능한 조합 생성
for _ in range(50):  # 예시로 100개 문장 생성
    # 첫 번째 리스트와 두 번째 리스트에서 요소 선택
    sentence_parts = []
    sentence_parts.append(random.choice(lists[0]) + random.choice(lists[1]))

    # 세 번째, 네 번째, 다섯 번째 리스트에서 랜덤하게 1개씩 선택하여 섞기
    part_indices = random.sample(range(2, 5), 3)
    random.shuffle(part_indices)
    for i in part_indices:
        sentence_parts.append(random.choice(lists[i]))

    # 여섯 번째 리스트에서 요소 선택
    sentence_parts.append(random.choice(lists[5]))

    # 문장 생성 및 저장
    sentence = ' '.join(sentence_parts)
    requirement_words.append(sentence)

# 결과 확인
for idx, sentence in enumerate(requirement_words, start=1):
    print(f"{idx}. {sentence}")
```

아래는 원래 하던 것처럼, A씨를 E씨로 바꾸고... 유사도를 낮추려 바꿀 단어 리스트를 이용해
실제로 단어를 바꾸는 과정입니다.
여기에는 'synonym.txt'라는 파일(좌측 바에 있음)을 사용하므로, 수정하고 싶으면 거기서 수정해주세요.
저장하고 돌리는 것은 필수

```python
current_date = datetime.now().strftime("%Y%m%d_%H%M")
```

```python
import random
import os
import shutil
from datetime import datetime

# 전역 변수로 output_folder 선언
output_folder = ""

# 대체어 딕셔너리를 읽어오는 함수
def read_synonyms(filename):
    synonyms_dict = {}
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            parts = line.split(':')
            if len(parts) == 2:
                word = parts[0].strip().strip("'")
                synonyms = parts[1].strip().strip('[]\n').replace("'", "").split(', ')
                synonyms_dict[word] = synonyms
    return synonyms_dict

# 입력 텍스트 파일 처리 함수
def replace_words_with_synonyms(text, synonyms_dict):
    for word, synonyms in synonyms_dict.items():
        if word in text:
            text = text.replace(word, random.choice(synonyms))
    return text

# 추가된 단어 치환 및 파일 생성 함수
def replace_and_generate_files(input_file, base_word, requirement_words, synonyms_dict):
    global output_folder  # 전역 변수로 선언
    
    # 현재 날짜와 시간을 포맷에 맞게 생성
    current_date = datetime.now().strftime("%Y%m%d_%H%M")
    
    # 결과를 저장할 폴더 생성
    output_folder = os.path.join(base_word, current_date)
    os.makedirs(output_folder, exist_ok=True)
    
    # 각 requirement_word 별로 파일 생성
    for i, replacement in enumerate(requirement_words):
        modified_text = input_file.replace(base_word, replacement)
        modified_text = replace_words_with_synonyms(modified_text, synonyms_dict)
        
        # 파일명에 순서 번호를 붙여 겹치지 않게 만듦
        output_filename = os.path.join(output_folder, f"{i + 1}_{replacement}.txt")
        with open(output_filename, 'w', encoding='utf-8') as file:
            file.write(modified_text)
        print(f"{output_filename} 파일이 생성되었습니다.")

# 함수 정의: 원고 파일을 각 수신자에게 배분하는 함수
def distribute_files_to_recipients(input_folder, recipients, base_word, current_date):
    # 배분 기본 폴더
    base_folder = f'{base_word} 배분'

    # 각 수신자 폴더 경로 생성
    for recipient in recipients:
        recipient_folder = os.path.join(base_folder, current_date, recipient)
        os.makedirs(recipient_folder, exist_ok=True)

    file_list = sorted(os.listdir(input_folder))

    num_recipients = len(recipients)
    files_per_recipient = len(file_list) // num_recipients

    for i, file_name in enumerate(file_list):
        if file_name.endswith('.txt'):
            src = os.path.join(input_folder, file_name)
            recipient_index = i % num_recipients
            recipient = recipients[recipient_index]
            dst = os.path.join(base_folder, current_date, recipient, file_name)
            shutil.copy(src, dst)
            print(f"{file_name} 파일이 {recipient} 폴더에 복사되었습니다.")

# 대체어 딕셔너리 읽기
synonyms_dict = read_synonyms('synonym.txt')

recipients = ['이사님', '팀장님', '주휘']

# 파일 생성 함수 호출
replace_and_generate_files(input_file, base_word, requirement_words, synonyms_dict)

# 생성된 파일 목록 출력
if output_folder:
    print(f"\n폴더 '{output_folder}'의 내용:")
    for filename in os.listdir(output_folder):
        print(filename)

# 파일 배분 함수 호출
input_folder = rf'C:\Users\user\Desktop\원고치환\{output_folder}'
distribute_files_to_recipients(input_folder, recipients, base_word, datetime.now().strftime("%Y%m%d_%H%M"))

print("\n모든 파일 생성 및 배분이 완료되었습니다.")
```

여기서 또 돌리게 되면, 'synonym'을 이용해서 치환한 파일들이 저장됩니다. 
아까 설정했던 base_word인 '법인회생' 폴더 안에, '오늘날짜 및 시간' 폴더 안에 전체가 저장되었습니다.
